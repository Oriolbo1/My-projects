Since our objective is to determine is a game will be successful or not, this will require that we have a varaible that tells us  in a categorical way, how comertially succesfull a certain game is in steam. 

Currently, there is no varaible that give us this information, nontheles, we can determine that the more copies a game has sld, the more comertially successful it is. We do have a variable named "Number_of_game_owners" that tell us the number of units sold and thus we will use it to create our "Y" varaible, we will name it "Comercial_Performance".



To do that I will take the users that got the game (paying or for free) and will divide the quantities in the following quantiles "25%", "50%", "75%". Those games that have a number of players lower than the first quantile (25%) will be tagged as 'Commercial Failure'. The games that have a number of players between the first (25%) and second (50%) quantile will be tagged as 'Moderate Commercial Failure'. The games that are between the second and third will receibe the categorization of "Moderate Commercial Success" and finally the one that exced the 3rd quantile will be considered "Commercial Success".

Since in order to create a good machine learning model we need to have a balanced variable "Y", w will also check that the percentage of regsters for every category are similar


```python
data.rename(columns = {'SteamSpyOwners':'Number_of_game_owners'}, inplace = True)
  
data.loc[data['Number_of_game_owners'] <= data["Number_of_game_owners"].quantile(0.25), 'Comercial_Performance'] = 'Commercial Failure' 
data.loc[(data['Number_of_game_owners'] > data["Number_of_game_owners"].quantile(0.25)) & (data['Number_of_game_owners'] <= data["Number_of_game_owners"].quantile(0.5)), 'Comercial_Performance'] = 'Moderate Commercial Failure' 
data.loc[(data['Number_of_game_owners'] > data["Number_of_game_owners"].quantile(0.5)) & (data['Number_of_game_owners'] <= data["Number_of_game_owners"].quantile(0.75)), 'Comercial_Performance'] = "Moderate Commercial Success"
data.loc[data['Number_of_game_owners'] > data["Number_of_game_owners"].quantile(0.75), 'Comercial_Performance'] = "Commercial Success"
print("")

print(data.Comercial_Performance.value_counts(normalize=True)) #as we cab see here, each category represents 24-25% of the total. we can say that it is balanced. 
```

```python

#-------------------------------Data Science--------------------------------------

#The first thing that we will do is to select the model that we will be using, 
#To do this I took into consideration the models that could work the best for these case. 
#I took the following: 
    #-SVM:  thanks to the kernel "rbf" it can increase the dimensionality and separate datapoints that are difficult to divide.
            #I think that, since many games are similar in the dataset that we have, this model can be important doing the required divisions. 
    #-Random forest: This model give us a huge flexivility and since many games are similar in many columns but their market succes is drastically different,
            #is important to use a model that can diferenciate between critical parametters. Also the fact that it is and ensembled learning model, is super interesting for this case to reduce the error. 
    #-KNeighbors: This might be the model that is less adecuated for the kind of data that we have. Nonetheless, since a lot of the games that tend to use the same strategy have similar results, 
            #I thought that it migt be interestin to try out if it could be effective.


#The first thing is to use the GridSearch CV we create a code that will run trought different models using the different parammeters that we speciffy 
#I have choosen, for every model the parametter that, in my opinion, can affect more to the prodiction. 
#in this case, since we will be using the cros validation techniques (I define CV=5) I don't need to devide between test and training data.
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
model_params = {
    'svm': { 
        'model': svm.SVC(),
        'params' : { 
            'C': [1,10,20],            
            'kernel': ['rbf','linear'], 
            'gamma' :['scale','auto']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [5,10,50,100,200],
            'criterion':['gini','entropy']            
        }
    },

    'KNeighbors' : {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [5,10,20],
            'p': [1,2],
            'weights':['uniform','distance']
        }
    }
}



scores = []
for model_name, mp in model_params.items(): 
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)  
    clf.fit(X, Y) 
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_ 
    })
    
Modelos = pd.DataFrame(scores,columns=['model','best_score','best_params'])
pd.set_option("display.max_columns",20)
print(Modelos.sort_values(by="best_score",ascending=False))
```

```python
X_train ,X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25)
classifier = tf.keras.models.Sequential()
classifier.add(tf.keras.layers.Dense(units=200, activation='relu', input_shape=(142, )))
classifier.add(tf.keras.layers.Dropout(rate=0.4))
classifier.add(tf.keras.layers.Dense(units=400, activation='relu'))
classifier.add(tf.keras.layers.Dropout(rate=0.4))
classifier.add(tf.keras.layers.Dense(units=200, activation='relu'))
classifier.add(tf.keras.layers.Dropout(rate=0.4))
classifier.add(tf.keras.layers.Dense(units=4, activation='softmax'))

classifier.summary()
```

```python
classifier.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0013), loss='sparse_categorical_crossentropy', metrics = ['accuracy'])

epochs_hist = classifier.fit(X_train, Y_train, epochs = 15)


print(Y_test)
test_loss, test_acc = classifier.evaluate(X_test,  Y_test) 
print('Test accuracy:', test_acc)
```
